### 1. Создание и изменение массивов.
Создайте массив NumPy  размеров 3x3, содержащий числа от 1 до 9. Умножьте все элементы массива на 2, а затем замените все элементы больше 10 на 0. Выведите итоговый массив.

### 3. Объдинение и разбиение массивов.

### 4. Генерация и работа с линейными последовательностями.

```python
import numpy as np
A = np.linspace(-10, 10, 50)
print("Созданный массив:")
print(A)
B = np.sum(A)
print("Сумма всех элементов",B)
C = np.sum(A[A>0])
print("Сумма всех положительных элементов:",C)
D = np.sum(A[A<0])
print("Сумма всех отрицательных элементов:",D)

```

### 8. Определитель и обратная матрица.
Создайте случайную квадратную матрицу 3х3. Найдите и выведите:
- Определитель этой матрицы
- Обратную матрицу (если существует, иначе выведите сообщение, что матрица вырождена)

Используйте функции np.linalg.det и np.linalg.inv

```python
A = np.random.rand(3, 3)
print("Случайная квадратная матрица 3х3:")
print(A)
B = np.linalg.det(A)
print("Нашли определитель этой матрицы:")
print(B)
C = np.linalg.inv(A)
if (B != 0):
    print("Обратная матрица:")
    print(C)
else:
    print("Обратной матрицы не существует, она вырождена")
```

### 10. Системы линейных уравнений.
Решите систему линейных уравнений вида:

$
\begin{cases}
2x + 3y - z = 5 \\
4x - y + 2x = 6 \\
-3x + 5y + 4z = -2
\end{cases}
$




Используйте матричное представление Ax = B , где A – матрица коэффициентов, х - вектор неизвестных, В - вектор провой части. Решите систему с помощью np.linalg.solve и выведите результат.

### 11. Индивидуальное задание.
Решите индивидуальное задание согласно варианту. Каждое задание предусматривает построение системы линейных уравнений. Решите полученную систему уравнений с использованием библиотеки NumPy. Для решения системы используйте метод Крамера и матричный метод. Сравните полученные результаты, с результатами, полученными с помощью np.linalg.solve.

**Распределение нагрузки на работников.** Три сотрудника должны выполнить 100 задач. Второй сотрудник должен выполнить на 10 задач больше, чем первый, а третий - в два раза больше, чем второй. Сколько задач получит каждый сотрудник?

Получим систему уравнений:
$
\begin{cases}
x + y + z = 100 \\
-x + y = 10 \\
-2y + z = 0
\end{cases}
$

```python
import numpy as np

A = np.array([
    [1, 1, 1],    # x + y + z = 100
    [-1, 1, 0],    # -x + y = 10
    [0, -2, 1]     # -2y + z = 0
])
B = np.array([100, 10, 0])

n = np.linalg.solve(A, B)
print("Решение СЛУ с помощью np.linalg.solve:")
print(f"x = {n[0]:.0f}, y = {n[1]:.0f}, z = {n[2]:.0f}.")

G = np.linalg.inv(A)
m = G @ B
print("Решение СЛУ с помощью матричного метода:")
print(f"x = {m[0]:.0f}, y = {m[1]:.0f}, z = {m[2]:.0f}.")

det_A = np.linalg.det(A)

A_x = A.copy()
A_x[:, 0] = B
x = np.linalg.det(A_x) / det_A

A_y = A.copy()
A_y[:, 1] = B
y = np.linalg.det(A_y) / det_A

A_z = A.copy()
A_z[:, 2] = B
z = np.linalg.det(A_z) / det_A

k = np.array([x, y, z])
print("Решение СЛУ с помощью метода Крамера:")
print(f"x = {k[0]:.0f}, y = {k[1]:.0f}, z = {k[2]:.0f}.")
```
